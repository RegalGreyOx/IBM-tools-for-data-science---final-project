### This code will create a histogram of a normal distribution with a mean of 0 and a standard deviation of 1, using 1000 data points.

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate data for the normal distribution
data = np.random.normal(0, 1, 1000)

# Create a histogram
plt.hist(data, bins=50)

# Add labels
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Normal Distribution Histogram')
```
